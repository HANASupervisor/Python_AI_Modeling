```python
# classification
import joblib
import pandas as pd
import torch
import torch.nn as nn
```

```python
class Classifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(4, 10)
        self.leakyrelu = nn.LeakyReLU()
        self.fc2 = nn.Linear(10, 3)

    def forward(self, x):
        x = self.fc1(x)
        x = self.leakyrelu(x)
        x = self.fc2(x)
        return x
```

```python
DEVICE = 'cpu'
if torch.cuda.is_available():
    DEVICE = 'cuda'
```

```python
MODEL_PATH = './iris_model01.pth'

model = Classifier().to(DEVICE)
model.load_state_dict(torch.load(MODEL_PATH))
```

```python
SCALER_PATH = './std_scaler.joblib'
scaler = joblib.load(SCALER_PATH)
```

```python
df = pd.read_csv('./data/iris_without_y.csv', encoding='cp949')
```

```python
new_X_std = scaler.transform(df)
X_std = torch.tensor(new_X_std, dtype=torch.float32).to(DEVICE)
```

```python
model.eval()
with torch.no_grad():
    y_pred_probability = model.forward(X_std)
    y_pred_label = torch.argmax(y_pred_probability, 1)
```
