## **3. Machine learning 모델의 학습방법과 타겟데이터에 따른 구분**
**타겟데이터($y$)** 데이터의 존재 유무에 따라 **supervised leanring**(**지도학습**)과 **unsupervised**(**비지도 학습**)으로 구분하며 학습방법이 달라진다.    

### 3.1 **supervised learning**(**지도학습**)
- ML모델이 결과값을 출력하기 위한 정답/target값 까지 함께 학습하는 학습방법  
- 모델을 학습시키기 위해 $y$값(target data, 정답, 레이블)이 반드시 필요  
- **supervied learning** 학습방법 내 $y$값의 분포에 따라 - **prediction**(**예측**), **classification**(**분류**) 문제로 구분 할 수 있습니다.

#### 3.1.1 **prediction**(**예측**) : target data의 분포가 **continuous distribution**(**연속형분포**)를 보일 때
 - 연속형분포 데이터 - 매출, 온도, 시간, 코스피지수, 가격 등  
 - 자연수 0과 1사이에는 무수히 많은 실수가 존재하고 그 무한한 값의 범위를 갖는 데이터를 연속형 데이터라 한다.(실수데이터)

#### 3.1.2 **classification**(**분류**) : 종속변수가 **categorical distribution**(**범주형**)분포를 보일 때
* 범주형 데이터 - 성별, 신용등급, 브랜드, 기차역 등  
* 남성 혹은 여성 구분처럼 이진형 구분(0 혹은 1), 데이터가 특정 범위내에 모두 속한 데이터의 경우 범주형 데이터 혹은 카테고리 데이터라고 한다.

## **4. ML 프로젝트의 기본 흐름**
머신러닝 프로젝트는 일반적으로 다음의 단계를 거칩니다.

4. **평가**: 테스트 데이터를 활용하여 모델의 성능 측정.

5. **배포 및 모니터링**: 실제 서비스에 적용 후 성능 유지보수.

# Boston Housing 데이터 변수 설명

아래 데이터는 보스턴 지역의 주택 가격 예측에 자주 사용되는 **Boston Housing Dataset** 형식의 데이터입니다.  
각 행은 하나의 지역/구역을 의미하며, `y`는 해당 지역의 주택 가격을 나타내는 목표 변수입니다.

## 변수 설명표

| 변수명 | 의미 | 설명 |
|---|---|---|
| `CRIM` | 지역별 1인당 범죄율 | 해당 지역의 범죄율을 의미합니다. 값이 높을수록 범죄율이 높은 지역입니다. |
| `ZN` | 대형 주거지역 비율 | 25,000 평방피트 이상으로 구획된 주거용 토지의 비율입니다. 값이 높을수록 넓은 주거 구역이 많다는 의미입니다. |
| `INDUS` | 비소매 상업지역 비율 | 해당 지역에서 소매업이 아닌 상업/산업용 토지가 차지하는 비율입니다. 공장, 산업시설 등이 많은 지역일 가능성이 있습니다. |
| `CHAS` | 찰스강 인접 여부 | 찰스강과 인접한 지역인지 여부를 나타내는 더미 변수입니다. `1`이면 강과 인접, `0`이면 인접하지 않음을 의미합니다. |
| `NOX` | 일산화질소 농도 | 대기오염 정도를 나타내는 변수입니다. 값이 높을수록 공기 오염 수준이 높다고 볼 수 있습니다. |
| `RM` | 평균 방 개수 | 해당 지역 주택의 평균 방 개수입니다. 일반적으로 방 개수가 많을수록 주택 가격이 높아지는 경향이 있습니다. |
| `AGE` | 오래된 주택 비율 | 1940년 이전에 지어진 자가주택의 비율입니다. 값이 높을수록 오래된 주택이 많은 지역입니다. |
| `DIS` | 고용 중심지까지의 거리 | 보스턴의 주요 고용 중심지까지의 가중 거리입니다. 값이 클수록 고용 중심지와 멀리 떨어져 있음을 의미합니다. |
| `RAD` | 방사형 고속도로 접근성 지수 | 고속도로 접근성을 나타내는 지수입니다. 값이 높을수록 주요 도로 접근성이 좋다고 해석할 수 있습니다. |
| `TAX` | 재산세율 | 10,000달러당 재산세율을 의미합니다. 값이 높을수록 세금 부담이 큰 지역입니다. |
| `PTRATIO` | 학생-교사 비율 | 해당 지역의 학생 수 대비 교사 수 비율입니다. 값이 높을수록 교사 1명당 학생 수가 많다는 뜻입니다. 교육 환경과 관련된 변수로 볼 수 있습니다. |
| `B` | 인구 구성 관련 지표 | 원 데이터에서 특정 인구 비율을 변환해 만든 변수입니다. 현재는 윤리적 논란이 있어 해석과 사용에 주의가 필요한 변수입니다. |
| `LSTAT` | 저소득층 비율 | 해당 지역의 저소득층 인구 비율입니다. 값이 높을수록 저소득층 비중이 높은 지역입니다. |
| `y` | 주택 가격 | 목표 변수입니다. 해당 지역의 주택 가격 중앙값을 의미하며, 단위는 보통 1,000달러로 해석합니다. 예를 들어 `24.0`은 약 24,000달러를 의미합니다. |

## 분석 관점에서의 변수 구분

### 1. 목표 변수

| 변수명 | 설명 |
|---|---|
| `y` | 예측하고자 하는 주택 가격 |

### 2. 설명 변수

| 구분 | 변수 |
|---|---|
| 범죄/사회적 환경 | `CRIM`, `LSTAT`, `B` |
| 주거 및 토지 특성 | `ZN`, `INDUS`, `AGE` |
| 환경 요인 | `NOX`, `CHAS` |
| 주택 구조 | `RM` |
| 접근성 | `DIS`, `RAD` |
| 세금 및 교육 | `TAX`, `PTRATIO` |

## 해석 시 주의할 점

- `RM`은 평균 방 개수이므로 주택 가격과 양의 관계를 가질 가능성이 큽니다.
- `LSTAT`, `CRIM`, `NOX`는 일반적으로 주택 가격과 음의 관계를 가질 가능성이 있습니다.
- `CHAS`는 더미 변수이므로 `0`과 `1`로 해석해야 합니다.
- `RAD`, `TAX`는 지역 개발 수준이나 접근성과 관련될 수 있지만, 단순히 값이 높다고 주택 가격이 높다고 단정하면 안 됩니다.
- `B` 변수는 원 데이터셋의 역사적·윤리적 논란이 있는 변수이므로 분석 보고서에서는 사용 여부와 해석에 주의해야 합니다.

## **5.2 타겟데이터($y$) 분리**
- supervised learning은 정답데이터를 함께 학습하는 머신러닝 학습방법  
- 정답 데이터가 학습에 사용하는 설명변수에 포함 되어 있다면 머신러닝 모델은 정답데이터를 참고하여 모델링을 하기에 문제가 있을 수 있다.  
- 위에서 확인한 보스턴주택 데이터셋은 데이터셋 내 정답데이터로 사용 할 '$y$'변수가 포함되어 있어 이를 다른 변수명으로 구분하여 할당한다.  

<div align="center">
  <img src="https://raw.githubusercontent.com/overgroove/ML_lecture_image_data/main/image/07.png" height="300px" width="400px">
<div/>

```python
# X, y 분리
y = df['y']
X = df.drop('y', axis=1)
X
```

## **5.3 train test split**
- supervided 머신러닝 모델이 데이터를 학습하는데 있어 가장 핵심적인 과정이다.  
- 머신러닝 모델의 목적은 현재 사내에 축적되어 있는 데이터를 예측/분류하는 목적이 아님   
- 모수에 해당하는 신규데이터에 대한 자동화를 ML모델로 구현함에 있다.  
- 그렇기에 과거 데이터인 기존 데이터로 학습한 ML모델의 검증을 위한 검증용 데이터인 test데이터셋을 구분하는 과정을 거친다.  

<div align="center">
<img src="https://raw.githubusercontent.com/overgroove/ML_lecture_image_data/main/image/08.png" height="300px" width="400px">

```python
# 데이터셋 분리
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
X_train.shape, X_test.shape
```

## **5.4 모델생성**

<div align="center">
  <img src="https://raw.githubusercontent.com/overgroove/ML_lecture_image_data/main/image/09.png" height="300px" width="700px">
</div>

```python
# 모델 import
from sklearn.linear_model import LinearRegression
# 모델 정의
model = LinearRegression()
```

## **5.5 모델학습**
- 지도학습의 경우 X_train데이터에 대한 정답지에 해당하는 y_train데이터를 함께 모델에 전달하는 방법으로 모델 학습을 수행한다.  

<div align="center">
  <img src="https://raw.githubusercontent.com/overgroove/ML_lecture_image_data/main/image/10.png" height="300px" width="700px">
</div>

```python
# 모델 학습
model.fit(X_train, y_train)
```

## **5.6 모델 예측/분류 결과값 확인 및 모델 검증**
- 학습이 완료 된 모델은 모델이 학습한 결과값에 따라 예측/분류 결과값을 낼 수 있는 상태가 된다.  
- 학습이 완료 된 모델에 검증용 데이터로 구분해둔 test데이터셋의 정답데이터를 제외한 X_test데이터만을 전달 해 모델의 예측값인 y predicted data를 확인한다.  
- y predicted 출력 데이터의 정답데이터라 할 수 있는 y_test데이터를 우리는 갖고 있기에 이 두 데이터를 비교하여 모델을 객관적으로 평가 할 수 있다.  

<div align="center">
  <img src="https://raw.githubusercontent.com/overgroove/ML_lecture_image_data/main/image/11.png" height="300px" width="700px">
</div>

```python
# 모델 예측값 확인
pred = model.predict(X_test)
```

```python
# 모델 평가
from sklearn.metrics import r2_score
r2_score(y_test, pred)
```
